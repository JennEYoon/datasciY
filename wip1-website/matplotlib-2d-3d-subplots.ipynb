# MATLAB style subplots - plt


```python
# MATLAB style - plt interface
x = np.linspace(-5, 5, 30) 

plt.figure(figsize=(10, 5)) # create a plot figure
# create the first of two panels and set current axis
plt.subplot(1, 3, 1) # (rows, columns, panel number)
plt.plot(x, np.sin(x))

# create the second panel and set current axis
plt.subplot(1, 3, 2)
plt.plot(x, np.cos(x), 'b-')

# Add 3rd subplot
plt.subplot(1, 3, 3)
plt.plot(x, 2*np.sin(x), 'ro');
```

### 3D Contour Plots

```python
# Package for 3D graphs. 
# from mpl_toolkits import mplot3d

# Data for a contour graph.
def f(x, y):
    return np.sin(np.sqrt(x**2 + y**2)) # Sine wave applied to a circle.

x = np.linspace(-6, 6, 30)  # linearly spaced line, -6 to +6, 30 points.
y = np.linspace(-6, 6, 30)
X, Y = np.meshgrid(x, y)  # Convert 30 x and y points into a finer meshgrid.
Z = f(X, Y)  # Call f with a meshgrid X, Y inputs, get circular wave returned as a meshgrid.

fig6 = plt.figure(figsize=(10, 3.5))
ax1 = plt.axes(projection='3d')
ax1.contour3D(X, Y, Z, 50, cmap='binary')
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.set_zlabel('z')

# Increase viewing tilt and rotate 
ax1.view_init(60, 55);

plt.figure(figsize=(10, 5)) # create a plot figure
# create the first of two panels and set current axis
plt.subplot(1, 3, 1) # (rows, columns, panel number)
x = np.linspace(-5, 5, 30) 
plt.plot(x, np.sin(x))

# create the second panel and set current axis
plt.subplot(1, 3, 2)
plt.plot(x, np.cos(x));

# Add 3rd subplot
plt.subplot(1, 3, 3)
plt.plot(x, 2*np.sin(x))
```

```python
# MATLAB style - plt interface

plt.figure(figsize=(10, 5)) # create a plot figure
# create the first of two panels and set current axis
plt.subplot(1, 3, 1) # (rows, columns, panel number)
plt.plot(x, np.sin(x))

# create the second panel and set current axis
plt.subplot(1, 3, 2)
plt.plot(x, np.cos(x));

# Add 3rd subplot
plt.subplot(1, 3, 3)
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 50, cmap='binary')
```
