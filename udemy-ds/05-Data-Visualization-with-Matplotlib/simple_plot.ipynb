
# Simple Plot


Create a simple plot.



```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

# Data for plotting
t = np.arange(0.0, 2.0, 0.01)  # t is x-axis, 0 to 2, step .01.
s = 1 + np.sin(2 * np.pi * t)   # s is y-axis, sine wave.

fig, ax = plt.subplots()  # create figure and ax object.
ax.plot(t, s)  # x-axis first, y-axis next.

ax.set(xlabel='time (s)', ylabel='voltage (mV)',
       title='About as simple as it gets, folks')
ax.grid()

fig.savefig("test.png")
plt.show()
```
