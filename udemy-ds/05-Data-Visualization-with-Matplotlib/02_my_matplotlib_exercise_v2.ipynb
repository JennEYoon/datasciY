## Import Libraries

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
```

## Data Step
Create dummy data using random function of numpy.

```python
x = np.linspace(0, 5, 11)
y = x**2
```

```python
print(x, y)
```

## Functional Method -  plt.plot()

```python
# Functional method, plt.plot()
# x is x-axis scale, y is y-axis data.
plt.plot(x, y)
```

#### Change figure line color to red, tickmarks (circle, dash).

```python
plt.plot(x, y, 'ro-')
```

#### Add labels for x & y axis, Title, and edit line color/symbol.

```python
# Edit plot attributes.
plt.plot(x, y, 'ro-')  # Repeat here to define base plot.
plt.xlabel('x-axis, scale 0 to 5')
plt.ylabel('y-axis, scale 0 to 25')
plt.title('Figure 1: Simple Plot')
plt.show()
```

## Subplots Method, plt.subplots(rows, cols, graph id)

```python
# Function subplots()
# 1 rows, 2 columns, 1st column object.
plt.subplot(1, 2, 1)
plt.plot(x, y, 'ro-')

# 1 rows, 2 columns, 2nd column object.
plt.subplot(1, 2, 2)
plt.plot(x, y**.25, 'b-')
```

#### Add plot attributes.

```python
# Add subplot attributes.

# 1 rows, 2 columns, 1st column graph.
plt.subplot(1, 2, 1)
plt.plot(x, y, 'ro-')
plt.xlabel('x-axis, scale o to 4')
plt.ylabel('y-axis, scale o to 25')
plt.title('Figure 2 Left')

# 1 rows, 2 columns, 2nd column graph.
plt.subplot(1, 2, 2)
plt.plot(x, y**.25, 'b-')
plt.xlabel('x-axis, scale o to 4')
plt.ylabel('y-axis, scale 0 to 2')
plt.title('Figure 2 Right')

# Add space between left and right graphs, wspace for width between space.
# matplotlib.pyplot.subplots_adjust(left=None, bottom=None, right=None, top=None, wspace=None, hspace=None)
plt.subplots_adjust(top=0.92, bottom=0.08, left=0.10, right=0.95, hspace=0, wspace=0.35)
plt.show()
```

## Object Oriented Method - Basic Plot

```python
# Object oriented graphing.
# Create blank figure object, then call methods on the figure object.

fig = plt.figure()  # create a blank figure object

axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])  # define four corners (axes) of figure object

axes.plot(x, y)  # plot data
```

#### Add attributes to figure

```python
# Create figure object, and assign axes (4 corners).
fig = plt.figure() 
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])  

# Add attributes
axes.set_xlabel('x-axis, scale 0 to 5')
axes.set_ylabel('y-axis, scale 0 to 25')
axes.set_title('Figure 3, Object Oriented Method')

# plot data
axes.plot(x, y, 'b^-')  # plot data
```

## Object Oriented Method

#### OO Example 1, one graph inside another graph.

```python
# Define figure object
fig = plt.figure()

# Define axes for outer graph, 4-corners.
# from left, from bottom, width, height (range 0 to 1)
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]) 

# Define axes for inside graph.
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])
```

#### Add attributes and plot data

```python
# Creates blank canvas
fig = plt.figure()
# Add title to overall figure.
fig.suptitle('Figure 4 - OO Ex1, inset graph', fontsize=16)

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.7]) # main axes
axes2 = fig.add_axes([0.2, 0.4, 0.4, 0.3]) # inset axes

# Larger Figure Axes 1
axes1.plot(x, y, 'b^-')
axes1.set_xlabel('X_label_axes2')
axes1.set_ylabel('Y_label_axes2')
axes1.set_title('Axes 1 Title')

# Insert Figure Axes 2
axes2.plot(y, x, 'ro-')
axes2.set_xlabel('X_label_axes2')
axes2.set_ylabel('Y_label_axes2')
axes2.set_title('Axes 2 Title');  # ; suppres object printing in jupyter notebook.
```

#### OO Example 2, two graphs side-by-side

```python
# Define figure object
fig = plt.figure()

# Define axes for left graph.
# from left, from bottom, width, height (range 0 to 1)
axes1 = fig.add_axes([0.1, 0.1, 0.3, 0.7]) 
axes1.plot(x, y, 'b^-')

# Define axes for right graph.
axes2 = fig.add_axes([0.6, 0.1, 0.3, 0.7])
axes2.plot(y, x, 'ro-');  # ; used to supress object printing in notebook.
```

#### Add label attributes, figure title

```python
# Define figure object
fig = plt.figure()
fig.suptitle('Figure 5 - OO Ex2, two graphs', fontsize=16)

# Define axes for left graph.
# from left, from bottom, width, height (range 0 to 1)
axes1 = fig.add_axes([0.1, 0.1, 0.3, 0.7]) 
axes1.set_xlabel('X_label_axes1')
axes1.set_ylabel('Y_label_axes1')
axes1.set_title('Axes 1 Title')
axes1.plot(x, y, 'b^-')

# Define axes for right graph.
axes2 = fig.add_axes([0.6, 0.1, 0.3, 0.7])
axes2.set_xlabel('X_label_axes2')
axes2.set_ylabel('Y_label_axes2')
axes2.set_title('Axes 2 Title')
axes2.plot(y, x, 'ro-');  # ; used to supress object printing in notebook.
```

## Lesson 2, OO

### Object Oriented Method - Subplots

```python
# fig, axes = plt.subplot(nrows= , ncols= )
# Combine steps.  Instead of fig(), then axes = fig.add_axes().

fig, axes = plt.subplots(nrows=2, ncols=2)
# axes object is an array of subplot axis.
plt.tight_layout()  # add space between rows & columns.
```

#### axes index through array ( or can iterate through).

```python
# index through axes.
fig, axes = plt.subplots(nrows=1, ncols=2)

axes[0].plot(x, y, 'ro-')
axes[1].plot(y, x, 'b^-')
```

```python
# iterate through axes.
fig, axes = plt.subplots(nrows=2, ncols=1)

for ax in axes:
    ax.plot(x, y)
```

## Figure Size and DPI

```python
fig = plt.figure(figsize=(3, 2), dpi = 200)  # dpi=100 seems default.  200 is twice as large graph.

ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, y, 'bo-');
```

```python
# wider graph, (8, 3)
fig = plt.figure(figsize=(8, 3))

ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, y, 'bo-');
```

#### Subplot example

```python
# fig, axes = plt.subplots()
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(8, 5))

axes[0].plot(x, y, 'ro-')
axes[1].plot(y, x, 'b^-')
plt.tight_layout();
```

#### Add attributes

```python
# fig, axes = plt.subplots()
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(9, 5))
fig.suptitle('Figure 6 - Subplots', fontsize=16)

axes[0].plot(x, y, 'ro-')
axes[0].set_xlabel('axis1 xlabel')
axes[0].set_ylabel('axis1 ylabel')
axes[0].set_title('Title Axis1')

axes[1].plot(y, x, 'b^-')
axes[1].set_xlabel('axis2 xlabel')
axes[1].set_ylabel('axis2 ylabel')
axes[1].set_title('Title Axis2')

# with figure title, need to narrow axes further.
plt.tight_layout(rect=[0, 0.03, 1, 0.93])
```

```python
# saving to file with highter dpi.
fig.savefig('myfig6.png', dpi=200)
```

## Add Legend

```python
# Legend paramter to ax.plot(x, y, label='legend')

# Basic Method
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 0.85])
fig.suptitle('Figure 7 - Legend', fontsize=16)

ax.plot(x, x**2, 'bo-', label='X Squared')
ax.plot(x, x**3, 'r^-', label="X Cubed")
ax.legend()
```

```python
# saving to file with highter dpi.
fig.savefig('myfig7_legend.png', dpi=200)
```
