### 3D Contour Plots

```python
# Data for a contour graph.

def f(x, y):
    return np.sin(np.sqrt(x**2 + y**2)) # Sine wave applied to a circle.

x = np.linspace(-6, 6, 30)  # linearly spaced line, -6 to +6, 30 points.
y = np.linspace(-6, 6, 30)

X, Y = np.meshgrid(x, y)  # Convert 30 x and y points into a finer meshgrid.
Z = f(X, Y)  # Call f with a meshgrid X, Y inputs, get circular wave returned as a meshgrid.
```

```python
# Plotting - MatLab style

fig = plt.figure()  # Create a figure container.
ax = plt.axes(projection='3d')  # Create the default axes object.
ax.contour3D(X, Y, Z, 50, cmap='binary')  # 3D graph type is contour.

# ax.set_xlabel('x')
# ax.set_ylabel('y')
# ax.set_zlabel('z')

# Increase viewing tilt and rotate 
# ax.view_init(60, 35);
```

```python
fig3 = plt.figure()

ax1 = plt.axes(projection='3d')  # Create first axes ad 3D projection.
ax1.contour3D(X, Y, Z, 50, cmap='binary')

# x = np.linspace(-6, 6, 30)  # x-axis is linearly spaced 30 points from -6 to +6.
ax2 = plt.axes([0.65, 0.65, 0.2, 0.2])  # Create second axes as 2D.
plt.plot(x, np.sin(x))  # Plot 2D sine wave.

print(fig3, ax1, ax2)
```

# MATLAB style subplots - plt


```python
# MATLAB style - plt interface

fig4 = plt.figure(figsize=(10, 5)) # create a figure container.

# create the first axes
ax1 = plt.subplot(1, 3, 1) # (rows, columns, panel number)
plt.plot(x, np.sin(x))

# create the second axes
ax2 = plt.subplot(1, 3, 2)
plt.plot(x, np.cos(x));

# Add 3rd subplot
ax3 = plt.subplot(1, 3, 3)
plt.plot(x, 2*np.sin(x))

print(fig4, ax1, ax2, ax3)
```

```python
# Plotting - MatLab style
# Put 3D and 2D in one-row, subplots.

fig7 = plt.figure(figsize=(10, 5)) # create a plot figure

# create the first of two panels and set current axis
ax1 = plt.axes([0, 0, .55, .6],projection='3d')
ax1.contour3D(X, Y, Z, 50, cmap='binary')

ax2 = plt.axes([.7, .1, .4, .4])
plt.plot(x, np.sin(x))
```
