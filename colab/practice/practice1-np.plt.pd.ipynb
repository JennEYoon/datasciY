## 2. Plotting

```python
fig1 = plt.figure()
ax1 = plt.axes()
ax1.plot()
```

```python
fig, ax = plt.subplots(1, 3)
ax[0].plot(X1, np.sin(X1))
ax[1].plot(X3, np.sin(X3))
ax[2].plot(X2, X3)
```

```python
xval = np.linspace(-10, 10, 21)
yval = xval**2

xval2 = np.geomspace(1, 100, 21)
yval2 = xval2**(1/5)

xval3 = np.arange(1, 21)
yval3 = np.random.randint(-10, 11, size=20)
print(xval3, yval3)

fig, ax = plt.subplots(1, 3)
ax[0].plot(xval, yval)
ax[1].plot(xval2, yval2)
ax[2].plot(xval3, yval3, ':or')
```

```python
# replot with space between plots  
fig, ax = plt.subplots(1, 2)
ax[0].plot(xval, yval)
ax[1].plot(xval3, yval3, ':or')
plt.subplots_adjust(wspace=.5)
ax[0].set_title('x squared')
ax[1].set_title('random circles')
```
