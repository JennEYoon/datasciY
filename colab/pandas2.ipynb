```python
# Quick exercise, pandas library  
# Jan 10, 2022 6:40 am  
# Jan 13, 2022 3pm, run on Colab test  

import numpy as np
import pandas as pd  
import random  
```

```python
#Try with data = []
names = ["helen", "mary", "jane", "tom", "philippe", "craig"]
ids = np.arange(1000, 1006, 1)
df2 = pd.DataFrame(data=[names, ids])
print(df2.head(6))
# Surprosingly that worked, but was transposed.  
```

```python
# Transpose method  
# ex:  data = pd.DataFrame(data=[x,y]).transpose()  
# df2a = pd.DataFrame(data=[ids, names]).transpose() 
df2a = pd.DataFrame([ids, names]).T
df2a.columns = ['ids', 'names']
df2a.head(6)
# OK, that worked! Unexpected, no errors. 
```

```python
# Transpose the data first, then add to DataFrame.  
# Series class transpose to row-length already.  
# names2 = names.transpose()
# print(names2)
# error, can't transpose list.  
names2 = pd.Series(names)
ids2 = pd.Series(ids)
df2b = pd.DataFrame(data=[ids2, names2])
df2b.head(6)
#OK that didn't transpose it automatically  
```

```python
# Transpose after defining Series.  
ids3 = pd.Series(ids).T
names3 = pd.Series(names).T
print(ids3, names3)
# Alright, that worked.  
```

```python
# Now define DataFrame with transposed Series.  
ids3 = pd.Series(ids).T
names3 = pd.Series(names).T
df2c = pd.DataFrame(data=[ids3, names3])
df2c.head(6)
# That was unexpected, still didn't work.  
# DataFrame automatically reads in Series as rows.  
# No way around it.  First readin all columns as row, then transpose entire DF.  
# Then add column attribute.  
```

```python
# Try trnaspose as data is read into DataFrame  
df2d = pd.DataFrame(data=[ids]).T  # one column, transposed list. 
print(df2d)
df2d.columns = ['ids']  # add column name attribute
print(df2d)
df2d['names'] = names  # add 2nd column as list, keeps column format.  
df2d.head(6)
# Wow, that worked.  
# Once DF already exists in columns, can add a column using list.
```

```python
# try creating df by using one col, then adding second col.     
orderids = np.arange(10, 16, 1)
companies = ["apple", "nveddia", "google", "microsoft", "tesla", "netflix"]
df3 = pd.DataFrame(data=companies, columns=['Companies'])
### Why did this work? data in column not row.  
# Test more
print("df3 \n", df3.head(6))

# define a new df column, auto-append to end.  
df3['OrderIDs'] = orderids
print("\n df3 append \n", df3.head())

# Add price column
price = [10.3, 21.5, 33.0, 15.2, 8.21, 32.6]
df3['Price'] = price
print('\n df3 append \n', df3)
```

```python
# Easy way to change order of columns in dataframe? reindex  
# df = df.reindex(columns=column_names) in a list  
df4 = df3.reindex(columns=['OrderIDs', 'Companies', 'Price'])
df4.head(6)
# Yeah!  That worked! :-D  
```

```python
# Go back to df3, unchanged order?  
df3.head(6)  
# yes.  
```

```python
# Use insertion method, can specify column location, not only at end.  
# ex: df.insert(2, "Age", [21, 23, 24, 21], True)

df3.insert(1, "Price", [10.3, 21.5, 33.0, 15.2, 8.21, 32.6], True)
print("df3 insert\n", df3.head(6))
```
