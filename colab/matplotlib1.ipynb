```python
# imports, alphabetical order by module alias
import matplotlib as mpl
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt  

%matplotlib inline
```

```python
# Create data, Y is sine wave data, X is linear spacing. 
X = np.linspace(0, 6.283, 30)
Y = np.sin(X)
print("X: ", X, "\nY: ", Y)
```

```python
# Format of numbers difficult to read. Reformat to 3 decimals. 
X1 = np.zeros([1, 30])
np.around(X, decimals=3, out=X1)  
# 'a'round for full array.  Slightly inaccurate but fast. 
#  output to new array, so original significant digits are kept.  
print("X1: ", X1)
print("X: ", X)  # unchanged digits? Yes. 
```

```python
# For single number, np.format_float_positional(x, precision=n).
# np.round(array)  same as np.around(array) 
X2 = np.zeros([1, 30])
np.round(X, decimals=3, out=X2)
print(X2, X)
```

```python
# Pandas quick plot
s1 = pd.Series(data=Y)  
# Don't need X index.  auto-index X axis.  
s1.plot()
#plt.show()  # worked.  
#plt.close()
```

```python
# matplotlib without setup, defaults assumed.  
plt.plot(X, Y, 'ro')
#plt.show()
#plt.close()
```

```python
# matplotlib with fig, ax setup
fig = plt.figure(2)
ax = plt.axes()
ax.set_title("Axis title, blue dash plot")
plt.plot(X, Y, 'b--')
#plt.show()
#plt.close()
```

```python
# subplots method, ax = plt.subplots(r, c)
fig, ax = plt.subplots()
ax.plot(X, Y, 'go')
ax.set_title('fig, ax defined, suplots method')
#plt.show()
#plt.close()
```

```python
# fig, ax method, multiple data series in one axis.
# ax = plt.axes(), define axis.      
fig = plt.figure(3)
ax = plt.axes()
ax.plot(X, Y, 'ko-')
ax.plot(X, np.cos(X), 'r--')
ax.plot(X, -1.25*Y, 'go')
ax.set_title('add plots to 1 axis method')
#plt.show()
#plt.close()
```

```python
# add_axes method  
# Creating two axes
# add_axes([xmin,ymin,dx,dy])
fig = plt.figure(6)
ax=fig.add_axes([0, .5,1, .4]) 
# left%, bottom%, width%, height% 
#plt.plot(X, Y, 'ko--') 
# plt.plot also works, current axis is default.  
ax.plot(X, Y, 'go')

ax1=fig.add_axes([0, 0, 1, .4])
#plt.plot(X, np.cos(X), 'r--')  
# plt.plot also works, current axis is default. 
ax1.plot(X, np.cos(X), 'bo--')

ax.set_title('ax0')
ax1.set_title('ax1')

fig.suptitle('add_axes method, to single figure')
#plt.show() # Yeah! Works!  
#plt.close()
```

### Other types of plots, 3D projection, polar-projectiong, histogram plot  

```python
# Add 3D projection, use ax.plot() method only. 
from mpl_toolkits import mplot3d

# 3D data  
def f(x, y):
    return np.cos(np.sqrt(x**2 + y**2))

x = np.linspace(-4, 4, 36)
y = np.linspace(-4, 4, 36)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)
```

```python
# OO Style, set plot
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 60, cmap='RdBu')
ax.set(xlabel='X', ylabel='Y', zlabel='Z', title='Contour Plot 3D')
```

```python
# Histogram, overplotting with alpha transparancy

# data for 3 histograms, randomly generated  
x1 = np.random.normal(0, 0.5, 200)
x2 = np.random.normal(-2, 1, 200)
x3 = np.random.normal(3, 1.6, 200)
```

```python
plt.style.use('seaborn-whitegrid')
kwargs = dict(histtype='stepfilled', alpha=0.4, density=True, bins=25)
plt.hist(x1, **kwargs)
plt.hist(x2, **kwargs)
plt.hist(x3, **kwargs)
plt.title('Three Histograms')
plt.xlabel('X data')
plt.ylabel('Y count');
```

```python
### Simple Plot, 
# https://matplotlib.org/3.5.1/gallery/lines_bars_and_markers/simple_plot.html  
import matplotlib.pyplot as plt
import numpy as np
# Data for plotting
t = np.arange(0.0, 2.0, 0.01)
s = 1 + np.sin(2 * np.pi * t)
fig, ax = plt.subplots()
ax.plot(t, s)
ax.set(xlabel='time (s)', ylabel='voltage (mV)',
       title='About as simple as it gets, folks')
ax.grid()
fig.savefig("simple_plot.png")
#plt.show()
```
